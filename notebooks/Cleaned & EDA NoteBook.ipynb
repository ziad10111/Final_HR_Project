```python
#import needed libraries 
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
# Load the data
file_path = r'C:\Data Science\Final_HR_Project\Final_HR_Project\Final_HR_Project\data\raw\HR-Employee-Attrition.csv'
df = pd.read_csv(file_path)
```

```python
df
```

```python
null_values = df.isnull().sum()
null_values
```

```python
# Remove duplicates
df_cleaned = df.drop_duplicates()
df_cleaned
```

```python
# Drop the "EmployeeCount" column
df_final = df_cleaned.drop(columns=["EmployeeCount"])

# Save the DataFrame to a new CSV file without the "EmployeeCount" column
df_final.to_csv(r'C:\Data Science\Final_HR_Project\Final_HR_Project\Final_HR_Project\data\processed\HR-Employee-Attrition-Cleaned.csv', index=False)
```

```python
# Ques_1 : Distribution of age among the employees
age_distribution = df_final['Age'].describe()
age_distribution
```

```python
# Qes _2 : Attrition rates across different departments
attrition_by_department = df_final.groupby('Department')['Attrition'].value_counts(normalize=True)
attrition_by_department
```

```python
# Qes _3 : Average monthly income per job role
average_income_by_role = df_final.groupby('JobRole')['MonthlyIncome'].mean().sort_values()
average_income_by_role
```

```python
# Qes 4 : Impact between Job Level and Monthly Income:
correlation_joblevel_income = df_final['JobLevel'].corr(df['MonthlyIncome'])
correlation_joblevel_income
```

```python
# Qes 5 : Impact between Years at Company and Years Since Last Promotion:
correlation_years_company_promotion = df_final['YearsAtCompany'].corr(df['YearsSinceLastPromotion'])
correlation_years_company_promotion
```
